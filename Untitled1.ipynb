```python
# 1. LIBRARY KYUN? Python ko matrix multiply nahi aati. Numpy = Maths ka ustad
import numpy as np
import matplotlib.pyplot as plt

# 2. DATA KYUN? Real data nahi hai. Apna banaya taake formula check ho sake
# X = Parhai ke ghante, y = Marks. Dekho relation: 2 * ghante = marks
X = np.array([1, 2, 3, 4, 5])   # Input
y = np.array([2, 4, 6, 8, 10])  # Output - Jawab hona chahiye m=2, c=0

# 3. m aur c KYUN 0? Kyun ke computer ko nahi pata. Pehla andaza 0 lagate hain
m = 0.0  # Slope ka andaza
c = 0.0  # Intercept ka andaza
learning_rate = 0.01  # Qadam kitna bara. Bara hoga to gir jayega, chota hoga to late pohanchega

# 4. TRAINING LOOP - YAHI AI KA DIMAAGH HAI
for i in range(1000):  # 1000 dafa practice karo
    # STEP A: Andaza lagao - Current m,c se marks predict karo
    y_pred = m * X + c   # Formula: y = mx + c. Yehi line sabse important hai

    # STEP B: Ghalti nikalo - Asal marks - Andaze wale marks
    error = y - y_pred   # Agar error = 0, matlab perfect andaza

    # STEP C: Dhalan nikalo - Ghalti kam karne ke liye m,c ko kidhar le jana hai
    # Ye 2 line "Gradient Descent" hain. Ghabrana nahi, bas yaad rakho: ghalti * X
    m_gradient = -2 * np.mean(error * X)
    c_gradient = -2 * np.mean(error)

    # STEP D: m,c ko update karo - Ghalti ki ulit taraf qadam lo
    m = m - learning_rate * m_gradient
    c = c - learning_rate * c_gradient

# 5. RESULT DEKHO
print("Final m =", round(m, 2))  # 2.0 aana chahiye
print("Final c =", round(c, 2))  # 0.0 aana chahiye

# 6. GRAPH KYUN? Taake aankh se dekho line fit hui ya nahi
plt.scatter(X, y, color='red', label='Asal Data')
plt.plot(X, m*X + c, label='AI ki Line')
plt.legend()
plt.show()
```
